## Helper functions

These functions will help us visualize 2D classification problems and summarize results.

```python

def plot_2d_data(X, y, title="Dataset"):
    plt.figure()
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30)
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()

def plot_decision_boundary(model, X, y, title="Decision boundary"):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )
    grid = np.c_[xx.ravel(), yy.ravel()]
    zz = model.predict(grid).reshape(xx.shape)

    plt.figure()
    plt.contourf(xx, yy, zz, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30)
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()

def summarize_classification_results(results):
    df = pd.DataFrame(results).sort_values("test_accuracy", ascending=False)
    return df.reset_index(drop=True)

def summarize_regression_results(results):
    df = pd.DataFrame(results).sort_values("test_rmse")
    return df.reset_index(drop=True)
```

# Part 1: Classification

We will use three synthetic classification datasets:

1. **Mostly linear dataset**
2. **Moons dataset** (nonlinear)
3. **Circles dataset** (strongly nonlinear)

These make it easy to see when linear methods struggle and when more flexible models help.

```python

# Dataset A: mostly linear
X_lin, y_lin = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    class_sep=1.5,
    random_state=42
)

# Dataset B: moons
X_moons, y_moons = make_moons(n_samples=500, noise=0.25, random_state=42)

# Dataset C: circles
X_circles, y_circles = make_circles(n_samples=500, noise=0.12, factor=0.4, random_state=42)

plot_2d_data(X_lin, y_lin, "Classification dataset A: mostly linear")
plot_2d_data(X_moons, y_moons, "Classification dataset B: moons")
plot_2d_data(X_circles, y_circles, "Classification dataset C: circles")
```

## Main takeaways

- Once you understand the **idea** behind a model, sklearn makes the implementation easy.
